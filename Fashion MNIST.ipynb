```
import tensorflow as tf
print(tf.__version__)
```

```
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.models import Model
```

```
# load in the data
# load in the data
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train),(x_test, y_test) = fashion_mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0 # scaling the data
print("x_train.shape:",x_train.shape)
```

```
# the data is only 2D
# convolution expects heigh x width x color
x_train = np.expand_dims(x_train, -1) # expand 1 dim,which is color
x_test = np.expand_dims(x_test, -1)
print(x_train.shape)
```

```
# no of classes
K = len(set(y_train)) # set returns unique labels
print("number of classes:", K)
```

```
# build the model using the functional API
i = Input(shape=x_train[0].shape)
x = Conv2D(32, (3,3), strides=2, activation='relu')(i) # using '32' kernels and each kernel is (3,3)
x = Conv2D(64, (3,3), strides=2, activation='relu')(x)
x = Conv2D(128, (3,3), strides=2, activation='relu')(x)
x = Flatten()(x)
x = Dropout(0.2)(x) # dropout rate is 20%,means each neuron's dropout percentage is 20%
x = Dense(512, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(K, activation='softmax')(x)

model = Model(i,x) # create a object of 'Model' class and parameter is i and x(input and output)
```

```
# compile and fit
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
r = model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=15)
```

```
# plot confustion matrix
from sklearn.metrics import confusion_matrix
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes, 
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)(:, np.newaxis)
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
    print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i,format(cm[i,j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i,j] > thresh else "black")
        plt.tight_layout()
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()
    p_test = model.predict(x_test).argmax(axis=1)
    cm = confusion_matrix(y_test,p_test)
    plot_confusion_matrix(cm, list(range(10)))
```

```
# show some misclassified examples
misclassified_idx = np.where(p_test != y_test)[0] # find the index where predicted is not same as label(p_test means predicted)
i = np.random.choice(misclassified_idx) # take the random value
plt.imshow(x_test[i], cmap='gray')
plt.title("True label: %s Predicted: %s" %(y_test[i], p_test[i]))
```
